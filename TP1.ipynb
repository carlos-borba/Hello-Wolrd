# Questão 1

Escreva uma função em Python que some todos os números ímpares de 1 até um dado N, inclusive. O número N deve ser obtido do usuário. Ao final, escreva o valor do resultado desta soma.


```python
#Escreva uma função em Python que some todos os números ímpares de 1 até um dado N, inclusive.
#O número N deve ser obtido do usuário. Ao final, escreva o valor do resultado desta soma.

n = int(input("Digite o valor de n: "))
        
impar = 1 # primeiro número ímpar
soma = 0  #inicio da soma
while impar <= n: #some os números ímpares
    soma += impar 
    impar += 2 # próximo ímpar
              
print(soma)#resultado final
```

```python
#Faça uma função em Python que receba do usuário a idade de uma pessoa em anos, meses e dias e retorne essa idade expressa em dias.
#Considere que todos os anos têm 365 dias.
ano = int(input("Diga a sua idade em anos: "))
meses = int(input("Diga o número do mês que nasceu: ")) 
dia = int(input("Digite o dia que nasceu: "))

dias_ano = (ano * 365)
dias_mes = (meses * 30)
dias_total = dias_ano + dias_mes + dia
print(f"Você viveu {dias_total} dias na sua vida")
```

# Questão 3

Escreva uma função em Python que calcule o fatorial de um dado número N usando um for. O fatorial de N=0 é um. O fatorial de N é (para N > 0): N x (N-1) x (N-2) x … x 3 x 2 x 1. Por exemplo, para N=5 o fatorial é: 5 x 4 x 3 x 2 x 1 = 120. Se N for negativo, exiba uma mensagem indicando que não é possível calcular seu fatorial.


```python
#screva uma função em Python que calcule o fatorial de um dado número N usando um for.
#O fatorial de N=0 é um. O fatorial de N é (para N > 0): N x (N-1) x (N-2) x … x 3 x 2 x 1.
#Por exemplo, para N=5 o fatorial é: 5 x 4 x 3 x 2 x 1 = 120. Se N for negativo, exiba uma mensagem indicando que não é possível calcular seu fatorial.

n = int(input("Fatorial de : "))#pedindo a entrada ao usuario

fatorial = 1 #onde começa toda a soma, ja que para fatorial de N=0 é um
if n < 0: #já testa se o número é negativo. 
    print("Não é possível calcular seu fatorial.") #caso seja negativo, ele indica ao usuário que nao pode calcular
else:
    for i in range(1, n): #para todo o resto, entra no for para o cálculo
        fatorial *= n 
        n -= 1 #seguindo a regra do fatorial: N x (N-1), até sair do range do for
    
    print(fatorial)
```

# Questão 4

Escreva um programa em Python que calcule o fatorial de um dado número N usando um while. Use as mesmas especificações do item anterior.

```python
#Escreva um programa em Python que calcule o fatorial de um dado número N usando um while. 
#Use as mesmas especificações do item anterior.
n = int(input("Fatorial de : "))

fatorial = 1
contador = 1

while contador < n:
  if n < 0:
    print("Não é possível calcular seu fatorial.")
    
  else:
    fatorial *= n
    n -= 1

print(fatorial)
```

Para que X, Y e Z formem um triângulo é necessário que a seguinte propriedade seja satisfeita: o comprimento de cada lado de um triângulo deve ser menor do que a soma do comprimento dos outros dois lados. Além disso, o programa deve identificar o tipo de triângulo formado observando as seguintes definições:

1. Triângulo Equilátero: os comprimentos dos três lados são iguais.
2. Triângulo Isósceles: os comprimentos de dois lados são iguais.
3. Triângulo Escaleno: os comprimentos dos três lados são diferentes.


```python
x = int(input("Primeiro valor: "))
y = int(input("Segundo valor: "))
z = int(input("Terceiro valor: "))

triangulo = (x, y, z)

if x < y + z and y < x + z and z < x + y:
    print("Esses valores podem formar um triângulo ", end='')
    if x == y == z:
        print("equilátero!")
    elif x != y != z != x:
        print("escaleno!")
    else:
        print("isósceles!")
else:
    print("Esses valores não podem formar um triângulo!")
```
